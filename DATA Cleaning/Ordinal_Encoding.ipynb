```python
 import pandas as pd
```

```python
df=pd.DataFrame({"Size":['s','m','l','xl','s','m','l','s','s','l','xl','m']})
df.head(5)
```

```python
ord_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
df["Size_en"]=oe.transform(df[["Size"]])
```

```python
df
```

```python
ord_data1={"s":0,"m":1,"l":2,"xl":3}
```

```python
df["Size_en_map"]=df["Size"].map(ord_data1) #we  can assign random numbers using map function
```
