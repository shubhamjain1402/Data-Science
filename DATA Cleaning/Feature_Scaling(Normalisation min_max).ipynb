```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv(r"C:\Users\shubh\Downloads\Coding\ML\loan.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"]=ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_min"])
plt.show()
```
